```python
import numpy as np
df['age'] = np.random.randint(18, 66, size = len(df))
df
```

```python
df_18_25 = df.query('age >= 18 & age <= 25')
df_18_25
```

```python
count_18_25 = df_18_25['id'].count()
av_age_18_25 = df_18_25['age'].mean()
print(f'Количество пользователей: {count_18_25}. Средний возраст: {av_age_18_25:.2f}')
```
