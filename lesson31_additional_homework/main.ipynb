```python
from confluent_kafka.admin import AdminClient, NewTopic

conf = {
    'bootstrap.servers': 'localhost:9094'
}

admin = AdminClient(conf)
```

```python
new_topic = NewTopic(
    topic='raw_data',
    num_partitions=3,
    replication_factor=1
)

fs = admin.create_topics([new_topic])

for topic, f in fs.items():
    try:
        f.result()
        print(f"[v] Топик {topic} создан")
    except:
        print("Ошибка")
```

```python
import json
import random
import time
from datetime import datetime, timezone
from confluent_kafka import Producer

conf = {
    'bootstrap.servers': 'localhost:9094'
}

producer = Producer(conf)
TOPIC = 'raw_data'

def generate_message():
    return {
        "type": random.choice(['temperature', 'humidity']),
        "timestamp": datetime.now().isoformat(),
        "value": round(random.uniform(20.0, 35.0), 1)
    }

try:
    print("[v] Продюсер начал работать..")
    while True:
        msg = generate_message()
        producer.produce(TOPIC, value=json.dumps(msg))
        producer.poll(0)
        time.sleep(0.01)

        print('[v] Отправил 1 сообщение', datetime.now().isoformat())

except KeyboardInterrupt:
    print("[x] Продюсер остановлен")

finally:
    producer.flush()
```
